The analysis of the concurrent_users variable shows that NiTPHLeX recorded 3,487 observations, with an average of about 898 users online simultaneously. However, the high standard deviation of 663 indicates large fluctuations in user activity, ranging from as low as 200 users during off-peak hours to as high as 2,997 at peak demand. This wide gap between the mean and the maximum highlights that while the system usually handles moderate loads, extreme spikes in concurrent users are straining the CDN infrastructure, leading to issues such as buffering, long load times, and reduced performance during peak periods.
